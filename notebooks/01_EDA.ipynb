````python
# Cell 1 — Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (10, 5)

print("✅ Libraries imported successfully!")
````

````python
# Cell 2 — Load dataset
df = pd.read_csv('../data/creditcard.csv')

print("Shape:", df.shape)
print("\nFirst look:")
df.head()
````

````python
# Cell 3 — Check for nulls and data types
print("=== NULL VALUES ===")
print(df.isnull().sum())

print("\n=== DATA TYPES ===")
print(df.dtypes)

print("\n=== BASIC STATS ===")
df.describe()
````

````python
# Cell 4 — Class distribution (numbers)
print("=== CLASS DISTRIBUTION ===")
print(df['Class'].value_counts())
print("\nFraud Percentage:")
print(df['Class'].value_counts(normalize=True) * 100)


````

````python
# Cell 5 — Visualize class imbalance
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
sns.countplot(x='Class', data=df,palette=['#2ecc71', '#e74c3c'], ax=axes[0])
axes[0].set_title('Transaction Count: Legit vs Fraud', fontsize=14)
axes[0].set_xticklabels(['Legit (0)', 'Fraud (1)'])
axes[0].set_ylabel('Count')

# Add count labels on bars
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height()):,}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=12)

# Pie chart
sizes = df['Class'].value_counts()
axes[1].pie(sizes, labels=['Legit', 'Fraud'],
            autopct='%1.2f%%',
            colors=['#2ecc71', '#e74c3c'],
            startangle=90,
            explode=(0, 0.1))  # explode fraud slice
axes[1].set_title('Fraud vs Legit Percentage', fontsize=14)

plt.tight_layout()
plt.savefig('../data/class_distribution.png')  # save the plot
plt.show()
print("✅ Plot saved!")
````

````python
# Cell 6 — Separate fraud and legit
fraud = df[df['Class'] == 1]
legit = df[df['Class'] == 0]

print("Fraud transactions:", len(fraud))
print("Legit transactions:", len(legit))

print("\n=== FRAUD AMOUNT STATS ===")
print(fraud['Amount'].describe())

print("\n=== LEGIT AMOUNT STATS ===")
print(legit['Amount'].describe())
````

````python
# Cell 7 — Plot Amount distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Legit amounts
axes[0].hist(legit['Amount'], bins=50, color='#2ecc71', alpha=0.7)
axes[0].set_title('Legit Transaction Amounts', fontsize=13)
axes[0].set_xlabel('Amount (€)')
axes[0].set_ylabel('Count')
axes[0].set_xlim([0, 1000])  # zoom in, ignore outliers

# Fraud amounts
axes[1].hist(fraud['Amount'], bins=50, color='#e74c3c', alpha=0.7)
axes[1].set_title('Fraud Transaction Amounts', fontsize=13)
axes[1].set_xlabel('Amount (€)')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../data/amount_distribution.png')
plt.show()
````

````python
# Cell 8 — Convert Time to hours
df['hour'] = (df['Time'] // 3600) % 24

fraud_hour = df[df['Class'] == 1]
legit_hour = df[df['Class'] == 0]

# Plot fraud vs legit by hour of day
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

axes[0].hist(legit_hour['hour'], bins=24, color='#2ecc71',
             alpha=0.7, label='Legit')
axes[0].set_title('Legit Transactions by Hour of Day', fontsize=13)
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Count')

axes[1].hist(fraud_hour['hour'], bins=24, color='#e74c3c',
             alpha=0.7, label='Fraud')
axes[1].set_title('Fraud Transactions by Hour of Day', fontsize=13)
axes[1].set_xlabel('Hour')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../data/time_distribution.png')
plt.show()
````

````python
##Day 3 work


# Cell 12 — Reload everything (run this first every day)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (12, 5)

df = pd.read_csv('../data/creditcard.csv')
df['hour'] = (df['Time'] // 3600) % 24  # feature from Day 2

fraud = df[df['Class'] == 1]
legit = df[df['Class'] == 0]

print("✅ Data loaded!")
print(f"Fraud: {len(fraud)} | Legit: {len(legit)}")
# ```

# ---

## STEP 3 — KDE Plots for Top Features (20 mins)

### What is a KDE plot?
# KDE = **Kernel Density Estimation** — it's like a smooth histogram. Instead of bars, you get a smooth curve showing where values are concentrated.

# **Why use it here?**
# - Bar charts show counts
# - KDE shows the **shape of distribution**
# - When fraud and legit curves barely overlap → that feature is powerful for detection
# - When they overlap a lot → that feature is weak
# ```
# Powerful feature (V14):         Weak feature:
                                
# Legit  ███                      Legit  ████████
# Fraud      ███                  Fraud  ████████
#        ───────►                        ───────►
#   barely overlap = GOOD         heavily overlap = BAD
````

````python
# Cell 13 — KDE plots for top 6 features
top_features = ['V14', 'V17', 'V12', 'V10', 'V11', 'V16']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()  # makes it easier to loop

for i, feature in enumerate(top_features):
    # Plot legit distribution
    sns.kdeplot(data=legit[feature],
                ax=axes[i],
                label='Legit',
                color='#2ecc71',
                fill=True,
                alpha=0.4)

    # Plot fraud distribution
    sns.kdeplot(data=fraud[feature],
                ax=axes[i],
                label='Fraud',
                color='#e74c3c',
                fill=True,
                alpha=0.4)

    axes[i].set_title(f'{feature} Distribution: Fraud vs Legit',
                      fontsize=13)
    axes[i].legend()
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Density')

plt.suptitle('KDE Plots — Top Fraud Indicator Features',
             fontsize=16, y=1.02)
plt.tight_layout()
plt.savefig('../data/kde_top_features.png')
plt.show() 
````

````python
# Cell 14 — Amount KDE (log scale)
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Raw amount
sns.kdeplot(data=legit['Amount'], ax=axes[0],
            label='Legit', color='#2ecc71', fill=True, alpha=0.4)
sns.kdeplot(data=fraud['Amount'], ax=axes[0],
            label='Fraud', color='#e74c3c', fill=True, alpha=0.4)
axes[0].set_title('Amount Distribution (Raw)', fontsize=13)
axes[0].set_xlabel('Amount (€)')
axes[0].legend()
axes[0].set_xlim([0, 500])  # zoom in

# Log transformed amount
legit_log = np.log(legit['Amount'] + 1)
fraud_log = np.log(fraud['Amount'] + 1)

sns.kdeplot(data=legit_log, ax=axes[1],
            label='Legit', color='#2ecc71', fill=True, alpha=0.4)
sns.kdeplot(data=fraud_log, ax=axes[1],
            label='Fraud', color='#e74c3c', fill=True, alpha=0.4)
axes[1].set_title('Amount Distribution (Log Transformed)', fontsize=13)
axes[1].set_xlabel('Log(Amount + 1)')
axes[1].legend()

plt.suptitle('Transaction Amount: Raw vs Log Scale', fontsize=15)
plt.tight_layout()
plt.savefig('../data/amount_kde.png')
plt.show()
````

````python
# Cell 15 — Box plots to find outliers
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Amount outliers
sns.boxplot(x='Class', y='Amount', data=df,
            palette=['#2ecc71', '#e74c3c'], ax=axes[0])
axes[0].set_title('Amount Outliers', fontsize=13)
axes[0].set_xticklabels(['Legit', 'Fraud'])
axes[0].set_ylim([0, 5000])

# V14 outliers
sns.boxplot(x='Class', y='V14', data=df,
            palette=['#2ecc71', '#e74c3c'], ax=axes[1])
axes[1].set_title('V14 Outliers', fontsize=13)
axes[1].set_xticklabels(['Legit', 'Fraud'])

# V17 outliers
sns.boxplot(x='Class', y='V17', data=df,
            palette=['#2ecc71', '#e74c3c'], ax=axes[2])
axes[2].set_title('V17 Outliers', fontsize=13)
axes[2].set_xticklabels(['Legit', 'Fraud'])

plt.suptitle('Outlier Analysis', fontsize=15)
plt.tight_layout()
plt.savefig('../data/outlier_analysis.png')
plt.show()
````

````python
# Cell 16 — Count outliers using IQR method
def count_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return len(outliers)

print("=== OUTLIER COUNTS ===")
for col in ['Amount', 'V14', 'V17', 'V12']:
    print(f"{col}: {count_outliers(col)} outliers")
````

````python
# Cell 17 — Pairplot (use sample for speed)
# Take a balanced sample — 500 legit + all 492 fraud
legit_sample = legit.sample(500, random_state=42)
combined = pd.concat([legit_sample, fraud])

pairplot_features = ['V14', 'V17', 'V12', 'Amount', 'Class']

sns.pairplot(combined[pairplot_features],
             hue='Class',
             palette={0: '#2ecc71', 1: '#e74c3c'},
             plot_kws={'alpha': 0.5},
             diag_kind='kde')

plt.suptitle('Pairplot — Top Features vs Fraud/Legit',
             y=1.02, fontsize=15)
plt.savefig('../data/pairplot.png')
plt.show()
````

````python
# Cell 18 — Final summary: mean values fraud vs legit
feature_cols = [f'V{i}' for i in range(1, 29)] + ['Amount', 'hour']

fraud_means = fraud[feature_cols].mean()
legit_means = legit[feature_cols].mean()
diff = (fraud_means - legit_means).sort_values()

plt.figure(figsize=(14, 8))
colors = ['#e74c3c' if x < 0 else '#2ecc71' for x in diff]
diff.plot(kind='barh', color=colors)
plt.title('Feature Mean Difference: Fraud vs Legit\n(Red = Fraud lower | Green = Fraud higher)',
          fontsize=14)
plt.xlabel('Mean Difference (Fraud - Legit)')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('../data/feature_mean_diff.png')
plt.show()

print("\nTop features where fraud differs MOST from legit:")
print(diff.abs().sort_values(ascending=False).head(10))
````
